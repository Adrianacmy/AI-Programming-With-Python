```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
pokeman = pd.read_csv('pokemon.csv')
print(pokeman.shape)
pokeman.head(5)
```

```python
sb.countplot(data=pokeman, y='type_1')
```

```python
base_color = sb.color_palette()[1]
```

```python
sb.countplot(data=pokeman, x='type_2', color=base_color)
plt.xticks(rotation=90)
```

```python
# Absolute vs. Relative Frequency
# By default, seaborn's countplot function will summarize and plot the data in terms of absolute frequency, or pure counts
#One method of plotting the data in terms of relative frequency on a bar chart is to just relabel the counts axis in terms of proportions. 
n_points = pokeman.shape[0]

max_count = pokeman['type_1'].value_counts().max()
max_prop = max_count / n_points
# generate tick mark location and names
tick_props = np.arange(0, max_prop, 0.05)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]

# create the plot
base_color = sb.color_palette()[2]
sb.countplot(data=pokeman, x='type_1', color=base_color)
plt.xticks(rotation=90)
plt.yticks(tick_props*n_points, tick_names)
plt.ylabel('proportion')
```

```python
# Rather than plotting the data on a relative frequency scale, you might use text annotations to label the frequencies on bars instead

base_color = sb.color_palette()[2]
sb.countplot(data=pokeman, x='type_1', color=base_color)
n_points = pokeman.shape[0]
type_counts = pokeman['type_1'].value_counts()
locs, labels = plt.xticks()
for loc, label in zip(locs, labels):
    count = type_counts[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/n_points)
    plt.text(loc, count-8, pct_string, ha='center', color='w')
plt.xticks(rotation=90)
```

## Counting missing Data

```python
df = pd.read_csv('pokemon.csv')
df.head(5)
```

```python
na_counts = df.isna().sum()
base_color = sb.color_palette()[2]
sb.barplot(na_counts.index.values, na_counts, color=base_color)
plt.xticks(rotation=90)
```

```python
# pie charts
sorted_counts = df['type_1'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 0,
        counterclock = True);
plt.axis('square') # this will make the results a circle
```

```python
# donuts charts
sorted_counts = df['type_1'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 0,
        counterclock = True, wedgeprops = {'width' : 0.4});
plt.axis('square') # this will make the results a circle
```

## Histograms

```python
bin_edges = np.arange(0, df['speed'].max()+5, 5);
plt.hist(data = df, x = 'speed', bins = bin_edges);
```

## Subplots

```python
plt.figure(figsize=[10,5])
plt.subplot(1,2,1)
bin_edges = np.arange(0,df['speed'].max() + 5, 5);
plt.hist(data=df, x='speed', bins=bin_edges);
# hist on right
plt.subplot(1,2,2)
bin_edges = np.arange(0, df['speed'].max() +1, 1);
plt.hist(data=df, x='speed', bins=bin_edges);
```

## Alternative Approach

```python
sb.distplot(df['speed'])
#The distplot function has built-in rules for specifying histogram bins, and by default plots a kernel density estimate (KDE) on top of the data.
#  The vertical axis is based on the KDE, rather than the histogram: you shouldn't expect the total heights of the bars to equal 1, but the area under the curve should equal 1
```

```python
bin_edges = np.arange(0, df['speed'].max()+5, 5)
sb.distplot(df['speed'], bins = bin_edges, kde = False,
            hist_kws = {'alpha' : 1})
# This approach is not necessary, it can be simply done by hist
```

## Choosing a Plot for Discrete Data
## Descriptive Statistics, Outliers, and Axis Limits

```python
plt.figure(figsize = [10, 5])

# histogram on left: full data
plt.subplot(1, 2, 1)
bin_edges = np.arange(0, df['weight'].max()+5, 5)
plt.hist(data = df, x = 'weight', bins = bin_edges)

# histogram on right: focus in on bulk of data < 35
plt.subplot(1, 2, 2)
bin_edges = np.arange(0, 35+1, 1)
plt.hist(data = df, x = 'weight', bins = bin_edges)
plt.xlim(0, 35) # could also be called as plt.xlim((0, 35))
```
